### âš¡ ç·´ç¿’å•é¡Œ 1

è¾ºã®é•·ã•ãŒãã‚Œãã‚Œæ¬¡ã®ã‚ˆã†ãªä¸‰è§’å½¢ã®é¢ç©ã‚’æ±‚ã‚ã¦ã¿ã‚ˆã†ã€‚

- è¾ºã®é•·ã•a: 5
- è¾ºã®é•·ã•b: 4
- è¾ºã®é•·ã•c: 3

> ğŸ“¨ **ãƒ’ãƒ³ãƒˆ**
>
> 3è¾ºï¼ˆ`a`, `b`, `c`ï¼‰ã®é•·ã•ã‹ã‚‰ä¸‰è§’å½¢ã®é¢ç©`area`ã‚’æ±‚ã‚ã‚‹ã«ã¯ã€Œãƒ˜ãƒ­ãƒ³ã®å…¬å¼ã€ã‚’ç”¨ã„ã¾ã™ã€‚
>
> ãƒ˜ãƒ­ãƒ³ã®å…¬å¼
> 
> $s = \frac{(a+b+c)}{2}$
> 
> $area = \sqrt{s(s-a)(s-b)(s-c)}$

> ãƒ«ãƒ¼ãƒ—ã®åº¦ã«1ã¤å¤§ãã„æ•°ã®æ–‡å­—åˆ—ã‚’è¶³ã—ã¦ã„ãã¾ã™ãŒã€ã“ã®æ–‡å­—åˆ—ã¯ `nonce`ã¨ã„ã†åå‰ã®å¤‰æ•°ã«æ ¼ç´ã™ã‚‹ã¨è‰¯ã„ã§ã—ã‚‡ã†ã€‚ã¾ãŸã€ãã®éš›ã€`str`é–¢æ•°ã‚’ç”¨ã„ã¦æ•°å€¤ã‚’æ–‡å­—åˆ—ã«å¤‰æ›ã—ã¦ã‹ã‚‰æ°åã®æ–‡å­—åˆ—ã«é€£çµã™ã‚‹ã‚ˆã†ã«ã—ã¾ã—ã‚‡ã†ã€‚

````python
# ã“ã®ã‚»ãƒ«ã«ã‚³ãƒ¼ãƒ‰ã‚’æ›¸ã„ã¦å®Ÿè¡Œã—ã¾ã—ã‚‡ã†ã€‚

name = "é•·è°·éƒ¨é™½ä¸€éƒ"
i = 0
max = 100

while i < max:
    nonce = str(i)
    encoded = (name + nonce).encode("utf-8")
    hash = hashlib.md5(encoded).hexdigest()
    
    ### ã“ã“ã‹ã‚‰è‡ªåˆ†ã§ã‚³ãƒ¼ãƒ‰ã‚’æ›¸ã„ã¦å®Œæˆã•ã›ã¾ã—ã‚‡ã†
    if hash[0] == "0":
        print(name+nonce)
        print(hash)

    ### ã“ã“ã¾ã§è‡ªåˆ†ã§ã‚³ãƒ¼ãƒ‰ã‚’æ›¸ã„ã¦å®Œæˆã•ã›ã¾ã—ã‚‡ã†
    i += 1
````
